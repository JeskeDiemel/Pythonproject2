```python
import mesa
print("You are using mesa version: " + mesa.__version__)

from mesa import Agent, Model
import random
```

```python
class Klant(Agent):
    def __init__(self, model, man, leeftijd, inkomen, stad, statiebeleid):
        super().__init__(model)
        self.man = man
        self.leeftijd = leeftijd
        self.inkomen = inkomen
        self.stad = stad
        self.statiebeleid = statiebeleid

    def milieu(self):
        factor_geslacht = 0.2
        factor_leeftijd = 0.2
        factor_inkomen = 0.35
        factor_stad = 0.35

        milieubewustzijn = 0

        if self.man:
            gemiddelde_MB = 0.7
            standaard_deviatie = 0.2
            MB_man = max(1, int(random.normalvariate(gemiddelde_MB, standaard_deviatie)))
            milieubewustzijn += MB_man * factor_geslacht
        else:
            gemiddelde_MB = 0.8
            standaard_deviatie = 0.2
            MB_vrouw = max(1, int(random.normalvariate(gemiddelde_MB, standaard_deviatie)))
            milieubewustzijn += MB_vrouw * factor_geslacht

        if self.leeftijd < 30:
            gemiddelde_MB = 0.8
            standaard_deviatie = 0.2
            MB_onder30 = max(1, int(random.normalvariate(gemiddelde_MB, standaard_deviatie)))
            milieubewustzijn += MB_onder30 * factor_leeftijd
        else:
            gemiddelde_MB = 0.5
            standaard_deviatie = 0.2
            MB_boven30 = max(1, int(random.normalvariate(gemiddelde_MB, standaard_deviatie)))
            milieubewustzijn += MB_boven30 * factor_leeftijd

        if self.inkomen > 3000:
            gemiddelde_MB = 0.7
            standaard_deviatie = 0.2
            MB_hooginkomen = max(1, int(random.normalvariate(gemiddelde_MB, standaard_deviatie)))
            milieubewustzijn += MB_hooginkomen * factor_inkomen
        else:
            gemiddelde_MB = 0.4
            standaard_deviatie = 0.2
            MB_laaginkomen = max(1, int(random.normalvariate(gemiddelde_MB, standaard_deviatie)))
            milieubewustzijn += MB_laaginkomen * factor_inkomen

        if self.stad == True:
            gemiddelde_MB = 0.8
            standaart_deviatie = 0.2
            MB_stad = max(1, int(random.normalvariate(gemiddelde_MB, standaart_deviatie)))
            milieubewustzijn += MB_stad * factor_stad
        else:
            gemiddelde_MB = 0.5
            standaart_deviatie = 0.2
            MB_platteland = max(1, int(random.normalvariate(gemiddelde_MB, standaart_deviatie)))
            milieubewustzijn += MB_platteland * factor_stad

        return milieubewustzijn

    def geld(self):
        factor_geslacht = 0.2
        factor_leeftijd = 0.2
        factor_inkomen = 0.35
        factor_stad = 0.35

        geldbewustzijn = 0

        if self.man:
            gemiddelde_GB = 0.7
            standaard_deviatie = 0.2
            GB_man = max(1, int(random.normalvariate(gemiddelde_GB, standaard_deviatie)))
            geldbewustzijn += GB_man * factor_geslacht
        else:
            gemiddelde_GB = 0.8
            standaard_deviatie = 0.2
            GB_vrouw = max(1, int(random.normalvariate(gemiddelde_GB, standaard_deviatie)))
            geldbewustzijn += GB_vrouw * factor_geslacht

        if self.leeftijd < 30:
            gemiddelde_GB = 0.8
            standaard_deviatie = 0.2
            GB_onder30 = max(1, int(random.normalvariate(gemiddelde_GB, standaard_deviatie)))
            geldbewustzijn += GB_onder30 * factor_leeftijd
        else:
            gemiddelde_GB = 0.5
            standaard_deviatie = 0.2
            GB_boven30 = max(1, int(random.normalvariate(gemiddelde_GB, standaard_deviatie)))
            geldbewustzijn += GB_boven30 * factor_leeftijd

        if self.inkomen > 3000:
            gemiddelde_GB = 0.7
            standaard_deviatie = 0.2
            GB_hooginkomen = max(1, int(random.normalvariate(gemiddelde_GB, standaard_deviatie)))
            geldbewustzijn += GB_hooginkomen * factor_inkomen
        else:
            gemiddelde_GB = 0.4
            standaard_deviatie = 0.2
            GB_laaginkomen = max(1, int(random.normalvariate(gemiddelde_GB, standaard_deviatie)))
            geldbewustzijn += GB_laaginkomen * factor_inkomen

        if self.stad == True:
            gemiddelde_GB = 0.8
            standaard_deviatie = 0.2
            GB_stad = max(1, int(random.normalvariate(gemiddelde_GB, standaard_deviatie)))
            geldbewustzijn += GB_stad * factor_stad
        else:
            gemiddelde_GB = 0.5
            standaard_deviatie = 0.2
            GB_platteland = max(1, int(random.normalvariate(gemiddelde_GB, standaard_deviatie)))
            geldbewustzijn += GB_platteland * factor_stad

        return geldbewustzijn

    def vriendschap(self):
        ...

    def scheidingsgedrag(self):
        if self.milieu() > 0.3:
            self.model.gescheiden_dozen += 1
        else:
            self.model.niet_gescheiden_dozen += 1

    def retourgedrag(self):
        if self.geld() > 0.8:
            self.model.geretourneerde_kratten += 1
        else:
            self.model.niet_geretourneerde_kratten += 1

    def step(self):
        self.milieu()
        self.geld()

        if self.statiebeleid:
            self.retourgedrag()
        else:
            self.scheidingsgedrag()
```

```python
class HelloFresh(Model):
    def __init__(self, num_agents):
        super().__init__()
        self.num_agents = num_agents
        self.agents_list = []

        self.gescheiden_dozen = 0
        self.niet_gescheiden_dozen = 0
        self.geretourneerde_kratten = 0
        self.niet_geretourneerde_kratten = 0


        for i in range(self.num_agents):
            a = Klant(self, man=False, leeftijd=30, inkomen=3500, stad= True, statiebeleid=False)
            self.agents_list.append(a)


    def step(self):
        for agent in self.agents_list:
            agent.step()
        print ("Het aantal gescheiden dozen is: ", self.gescheiden_dozen)
        print ("Het aantal niet_gescheiden dozen is: ", self.niet_gescheiden_dozen)
        print ("Het aantal geretourneerde_kratten is: ", self.geretourneerde_kratten)
        print ("Het aantal niet_geretourneerde_kratten is: ", self.niet_geretourneerde_kratten)
```
